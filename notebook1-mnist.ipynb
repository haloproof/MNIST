```python
train_dataframe = pd.read_csv('/kaggle/input/digit-recognizer/train.csv')
test_dataframe = pd.read_csv('/kaggle/input/digit-recognizer/test.csv')
X_train, y_train = train_dataframe.iloc[:,-784:], train_dataframe.iloc[:,0]
X_test = test_dataframe
X_train.shape, y_train.shape, X_test.shape
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC


# Create an SVM classifier
svm_classifier = SVC(kernel="linear", C=1)

# Train the SVM classifier on the training data
svm_classifier.fit(X_train, y_train)

# Make predictions on the test data
y_pred = svm_classifier.predict(X_test)

output = pd.DataFrame({'ImageId': range(y_pred.shape[0]), 'Label': y_pred})
output.to_csv('submission.csv', index=False)
print("Your submission was successfully saved!")
```
